```python
import numpy as np
import matplotlib.pyplot as plt
class DataPlot:
    def Plot1():
        x=np.array([10,20,30,40,50,60])
        y=np.array([65,55,45,35,25,15])
        plt.plot(x,marker="D",mfc="red",mec="white",ms=10,color="purple",linewidth="3",linestyle="dashed")
        plt.plot(y,marker="H",mfc="green",mec="white",ms=10,color="yellow",linewidth="3",linestyle="dashed")
        plt.grid(color="orange")
        plt.show()
DataPlot.Plot1()
```
